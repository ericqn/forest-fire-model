```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from observation_functions import get_observations_given_queries, get_probabilities_given_queries, get_quartiles, display_histogram
```

```python
fires_df = pd.read_csv('data/forestfires.csv')
fires_df
```

```python
#FFMC
display_histogram(fires_df, 'FFMC')

#DMC
display_histogram(fires_df, 'DMC')

#DC
display_histogram(fires_df, 'DC')

#ISI
display_histogram(fires_df, 'ISI')

#Area
display_histogram(fires_df, 'area')

#Fire v No Fire
fires_df['fire'] = fires_df['area'].apply(lambda x: x>0)
plt.bar(fires_df['fire'].value_counts().index.astype(str), fires_df['fire'].value_counts())
plt.xlabel("Fire Category")
plt.ylabel("Count")
plt.title("Fire Occurrences")
plt.show()

#Month
plt.bar(fires_df['month'].value_counts().index, fires_df['month'].value_counts())
plt.xlabel("Month")
plt.ylabel("Count")
plt.title("Observations by Month")
plt.show()
```
